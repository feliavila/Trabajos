Se realiza un gráfico de la absorbancia en función de la longitud de onda para una muestra del flavonoide rutina

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from google.colab import drive
drive.mount('/content/drive')
df=pd.read_excel('/content/drive/MyDrive/2- Ciclo Superior/3°/9- Química Bioorgánica/Informes/TP4/Libro1.xlsx')
listas=df.to_numpy().transpose().tolist()
wavelen=listas[0]
absorb=listas[1]



```

```python
plt.plot(wavelen,absorb, color="green")
plt.xlabel("Longitud de onda/nm")
plt.ylabel("Intensidad de Absorbancia")
plt.yticks([])
plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)
plt.title("Curva espectral de rutina en UV-Vis")
plt.figure(figsize=(8.88, 5.75))
plt.show()
```

```python
x = np.array([2.1590e-2, 1.0795e-2, 5.3975e-3, 2.69875e-3,0])
y = np.array([0.656, 0.315, 0.135, 0.066, 0])

coeficientes = np.polyfit(x, y, 1)
Abs = coeficientes[0] * x + coeficientes[1]
y_mean = np.mean(y)
SS_total = np.sum((y - y_mean) ** 2)
SS_residual = np.sum((y - Abs) ** 2)
R2 = 1 - (SS_residual / SS_total)


plt.scatter(x, y, color='blue')
plt.plot(x, Abs, color='green',linestyle='--')
plt.xlabel('Concentración /(mg/mL)')
plt.ylabel('Absorbancia')
plt.title('Curva de calibración para la concentración de rutina')
plt.text(0.05, 0.95, f'Abs = {coeficientes[0]:.4f} (ml/mg) [Rutina] {coeficientes[1]:.4f}',
         transform=plt.gca().transAxes, fontsize=12, verticalalignment='top')
plt.text(0.05, 0.90, f'$R^2$ = {R2:.4f}',
         transform=plt.gca().transAxes, fontsize=12, verticalalignment='top')

plt.grid(True,linestyle='--', linewidth=0.5, alpha=0.7)
plt.figure(figsize=(8.88, 5.75))
plt.show()
```
