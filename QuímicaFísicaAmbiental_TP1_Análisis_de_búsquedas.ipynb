En este código se analizan bases de datos extraídas de GoogleTrends para ver las cantidades de búsquedas en temas relacionadas con cambio climático.

```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel("/content/drive/MyDrive/2- Ciclo Superior/3°/7- Química Física Ambiental B/Excels/búsquedas1.xlsx")

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].bar(df['Keyword'], df['Semana'], color='skyblue')
axes[0].set_xlabel('Keyword')
axes[0].set_ylabel('Last week results')
axes[0].set_xticklabels(df['Keyword'], rotation=45, ha='right')

axes[1].bar(df['Keyword'], df['Mes'], color='lightgreen')
axes[1].set_xlabel('Keyword')
axes[1].set_ylabel('Last month results')
axes[1].set_xticklabels(df['Keyword'], rotation=45, ha='right')

plt.tight_layout()
plt.show()


```

```python
df2= pd.read_excel("/content/drive/MyDrive/2- Ciclo Superior/3°/7- Química Física Ambiental B/Excels/búsquedas2.xlsx")
plt.figure(figsize=(15.15*2, 7.51*2))
plt.plot(df2['Mes'], df2['carbon emissions'],  color='b', label='carbon emissions', linewidth=5)
plt.plot(df2['Mes'],df2['greenhouse effect'], color='r', label='greenhouse effect', linewidth=5)
plt.plot(df2['Mes'],df2['carbon footprint'],   color='g', label='carbon footprint', linewidth=5)
plt.plot(df2['Mes'],df2['climate change'],   color='purple', label='climate change', linewidth=5)
plt.plot(df2['Mes'],df2['global warming'],  color='orange', label='global warming', linewidth=5)
plt.plot(df2['Mes'],df2['Paris Agreement'],  color='yellow', label='Paris Agreement', linewidth=5)
plt.xlabel('Year',fontsize=22)
plt.ylabel('Searchs',fontsize=22)
plt.legend(fontsize=22)

years = df2['Mes'].str.split('-').str[0].unique()
year_indices = df2['Mes'].str.split('-').str[0].drop_duplicates().index
plt.gca().set_xticks(year_indices)
plt.gca().set_xticklabels(years)


plt.tick_params(axis='both', which='major', labelsize=22)
plt.grid(color='lightgray', linestyle=':', linewidth=0.8)
plt.show()
```
