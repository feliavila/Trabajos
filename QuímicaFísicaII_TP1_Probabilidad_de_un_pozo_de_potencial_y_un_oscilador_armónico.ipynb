Si se ejecuta este código, se devuelve:
una imagen con las tres primeras funciones de onda para una partícula libre en una caja, otra imagen con el cuadrado de esas funciones de onda.
Luego pide un invervalo sobre el que se desea conocer la probabilidad para la función de onda con n=2 para finalmente devuelver una imagen con la integración sobre la función densidad de probabilidad para n=2

```python
#EJERCICIO 1.1 (tomamos n=1,2,3)
import numpy as np
import matplotlib.pyplot as plt


def psi1(x):
  return np.sqrt(2/4)*np.sin(1*np.pi*x/4)
def psi2(x):
  return np.sqrt(2/4)*np.sin(2*np.pi*x/4)
def psi3(x):
  return np.sqrt(2/4)*np.sin(3*np.pi*x/4)

x_values= np.linspace(0,4,1000)

y_values1=psi1(x_values)
y_values2=psi2(x_values)
y_values3=psi3(x_values)

y_values1sq=y_values1**2
y_values2sq=y_values2**2
y_values3sq=y_values3**2

plt.figure()
plt.plot(x_values,y_values1,label="$\psi_1$")
plt.plot(x_values,y_values2,label="$\psi_2$")
plt.plot(x_values,y_values3,label="$\psi_3$")
plt.legend()
plt.xlabel("$x$/nm")
plt.ylabel(r"$\psi(x) \, / \, \mathrm{nm}^{-1/2}$")
plt.title("Funciones de estado para partículas en un pozo de potencial")

plt.figure()
plt.plot(x_values,y_values1sq,label="|$\psi_1^2|$")
plt.plot(x_values,y_values2sq,label="|$\psi_2^2|$")
plt.plot(x_values,y_values3sq,label="|$\psi_3^2|$")
plt.legend()
plt.xlabel("$x$/nm")
plt.ylabel(r"$\psi(x)^2 \, / \, \mathrm{nm}^{-1}$")
plt.title("Densidad de probabilidad para partículas en un pozo de potencial")

plt.show()


#EJERCICIO 1.2 (tomamos n=2)
delta_x=4/1000
inicio=float(input("Ingrese el inicio de donde quiere calcular la probabilidad en nm: "))
fin=float(input("Ingrese el final de donde quiere calcular la probabilidad en nm: "))
y_integral=y_values2sq[int(inicio/delta_x):int(fin/delta_x)]
prob=0
for y in y_integral:
  prob+= y*delta_x
print("La probabilidad de encontrar la partícula entre ", inicio, "nm y ", fin, "nm es ", prob)
plt.figure()
plt.plot(x_values,y_values2sq)
plt.xlabel("$x$/nm")
plt.ylabel(r"$\psi(x)^2 \, / \, \mathrm{nm}^{-1}$")
plt.fill_between(x_values[int(inicio/delta_x):int(fin/delta_x)],y_integral,color="#b2fab4")
plt.title("Probabilidad de encontrar la partícula entre "+str(inicio)+"nm y "+str(fin)+"nm")
plt.show()

```

Luego el programa devuelve: una imagen con las funciones de onda para los osciladores con v=1, v=2 y v=3. Finalmente el programa predice la probabilidad de encontrar al oscilador en la región loca.


```python
#EJERCICIO 2.1
from scipy.constants import m_p, m_n, h, hbar, pi

protones1=int(input("Ingrese el número de protones del primer elemento: "))
neutrones1=int(input("Ingrese el número de neutrones del primer elemento: "))
protones2=int(input("Ingrese el número de protones del segundo elemento: "))
neutrones2=int(input("Ingrese el número de neutrones del segundo elemento: "))
mu=(((protones1+neutrones1)*(protones2+neutrones2))/(protones1+neutrones1+protones2+neutrones2))
k=float(input("Ingrese el valor de la constante k="))
alfa=np.sqrt(k*mu/hbar**2)

def psi4(x):
  return np.power(alfa/pi, 1/4)*np.exp((-alfa*x**2)/2)
def psi5(x):
  return np.power(4*alfa**3/pi, 1/4)*x*np.exp((-alfa*x**2)/2)
def psi6(x):
  return np.power(alfa/(4*pi), 1/4)*(2*alfa*x**2-1)*np.exp((-alfa*x**2)/2)

x_valores= np.linspace(-1e-17,1e-17,2000)
y_valores4=psi4(x_valores)
y_valores5=psi5(x_valores)
y_valores6=psi6(x_valores)

plt.figure()
plt.plot(x_valores,y_valores4,label="$\psi_0$")
plt.plot(x_valores,y_valores5,label="$\psi_1$")
plt.plot(x_valores,y_valores6,label="$\psi_2$")
plt.legend()
plt.xlabel("$x$/m")
plt.ylabel(r"$\psi(x) \, / \, \mathrm{m}^{-1/2}$")
plt.title("Funciones de onda para un oscilador armónico")
plt.show()

#EJERCICIO 2.2
xclas= np.sqrt((2/alfa)*((np.sqrt(k/mu)/(2*pi))+0.5))
print ("El valor máximo del desplazamiento de la posición de equilibrio permitido por la física clásica es", xclas,"m")
delta_x1=x_valores[1]-x_valores[0]
inicio1 = np.abs(x_valores - (-xclas)).argmin()
fin1 = np.abs(x_valores - xclas).argmin()
y_integral4=y_valores4**2
y_integral5=y_valores5**2
y_integral6=y_valores6**2
prob4=0
prob5=0
prob6=0
for r in range (inicio1,fin1):
  prob4+= y_integral4[r]*delta_x1
for r in range (inicio1,fin1):
  prob5+= y_integral5[r]*delta_x1
for r in range (inicio1,fin1):
  prob6+= y_integral6[r]*delta_x1

print("La probabilidad de encontrar en la región loca a un oscilador con la función de onda psi_0 es ", 1-prob4)
print("La probabilidad de encontrar en la región loca a un oscilador con la función de onda psi_1 es ", 1-prob5)
print("La probabilidad de encontrar en la región loca a un oscilador con la función de onda psi_2 es ", 1-prob6)
```

Si se ejecuta este código pide el número de estado vibracional de onda de oscilador que se desea graficar, luego grafíca la densidad de probabilidad de ese estado y la densidad de probabilidad del oscilador clásico.
La función de poliniomios de Hermite utilizada no aguanta polinomios mucho más grandes que el décimo.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import hermite
import math

scale = 5

k = 410 # N/m
mu = 1.65e-27 # kg

nu = np.sqrt(k/mu)/(2*np.pi)
hbar = 1.054571817e-34 # J.s
alpha = (mu*k/hbar**2)**.25 # 1/m
xclas = np.sqrt(2/alpha**2*(nu+.5))

row, col = 1, 2
plt.figure(figsize=(col*scale, .75*row*scale))

plt.subplot(row, col, 1)
x0 = 1.1*xclas
x = np.linspace(-x0, x0, 5000)
pclas = 1/(xclas*np.pi) * 1/np.sqrt(np.clip(1 - (x/xclas)**2, 0, None))
plt.plot(x, pclas, c='k', label='clásico')
print([np.nansum([pclas[i]*(x[1]-x[0]) for i in range(len(x))])])
plt.title("Distribución de probabilidad de un oscilador clásico")
plt.xlabel(r"$x \, / \, \mathrm{m}$")
plt.ylabel(r"Densidad de probabilidad $m^{-1}$")

plt.subplot(row, col, 2)
x0 = 2e-10
x = np.linspace(-x0, x0, 5000)
for n in range(0, 100, 20):
    H = hermite(n)(alpha*x)
    N = 1/(2**n*math.factorial(n))**.5*np.sqrt(alpha)/np.pi**.25
    den = N * H * np.exp(-.5*(alpha*x)**2)
    den2 = den**2
    plt.plot(x, den2, label='n={}'.format(n))
    print([np.nansum([den2[i]*(x[1]-x[0]) for i in range(len(x))])])
plt.legend()
plt.title("Densidad de probabilidad para un oscilador cuántico.")
plt.xlabel(r"$x \, / \, \mathrm{m}$")
plt.ylabel(r"$\psi(x)^2 \, / \, \mathrm{m}^{-1}$")

# Ajustar espaciado entre gráficos
plt.subplots_adjust(wspace=0.35)  # Ajusta el espaciado horizontal entre los gráficos

plt.show()
```
